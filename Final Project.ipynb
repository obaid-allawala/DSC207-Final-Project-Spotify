```python
# Next step I want to preform some changes to the dataset, 
# (1) I will drop some columns
# (2) I see that the artist name is inside, " ['xxx']"  I want to remove those parentheses. 
# (3) I will reduce year to songs between 2016 and 2020
```

```python
# I want to reduce the size of the dataset, I will first drop columns that I have no intentions of using
df = spotify.drop(['id', 'album_id','artist_ids','disc_number','track_number','explicit','key','mode','time_signature','release_date'], axis=1)
df.head()
#perfect
```

```python
#we need to remove "'['" and "']'" from aritst column
```

```python
df['artists'] = df.artists.str.strip("'['")
df['artists'] = df.artists.str.strip("']'")
```

```python
# now lets reduce the year to songs between 2000 & latest and look at the shape, its still big
df.loc[df['year'] >= 2000].shape
```

```python
# better now, we removed songs uptil 2015
df = df.loc[df['year'] >= 2016]
df.shape
```

```python
#I want to change duration_ms's millisecond to minutes
#ill change the variable first then the column name
df["duration_ms"] = (df["duration_ms"] / 1000).round()
```

```python
df = df.rename(columns={'duration_ms': 'duration'})
#looks perfect
```
